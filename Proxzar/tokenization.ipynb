```python
def tokenize_images(data):
    # Rearrange image tensor to match the format expected by the model
    data = rearrange(data, 'b c h w -> b (c h w)') # Image patches into a sequence
    tokens = torch.tensor(data)
    return tokens
```

```python
# Loop through the batches
num_batches = 922
batch_size = 32
```

```python
for batch_idx in range(num_batches):
    batch_path = f"/home/sklaptop/Downloads/OneDrive_2023-12-18/P25 - Visual Search/preprocessed_image_tensors/preprocessed_batch_{batch_idx}.npy"
    preprocessed_data = np.load(batch_path)

    # Tokenize the images for the current batch
    tokenized_images = tokenize_images(preprocessed_data)

    # Save the tokenized images
    torch.save(tokenized_images, f"/home/sklaptop/Downloads/OneDrive_2023-12-18/P25 - Visual Search/tokenized_images/tokenized_images_{batch_idx}.pt")

    print(f"Tokenization of the image for {batch_idx} is complete")
```
