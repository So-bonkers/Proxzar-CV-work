```python
import os
import pandas as pd
```

```python
dataset_path = r"C:\Users\Administrator\Downloads\beacon_images_raw"
```

```python
categories =[]
subtypes=[]
fileid=[]
filenames =[]
```

```python
for filename in os.listdir(dataset_path):
    if filename.endswith(".jpg"):
        file_parts = filename.split("_")
        if len(file_parts) >=2:
            category = file_parts[0]
            subtype = file_parts[1]
            fileids = file_parts[2]
            fileid.append(fileids)
            subtypes.append(subtype)
            categories.append(category)
            filenames.append(filename)
```

```python
fileid = [fileid.replace(".jpg", "") for fileid in fileid]
fileid
```

```python
total_unique_categories = len(set(categories))
print(total_unique_categories)
```

```python
data = {"Category": categories, "Sub-Category": subtypes, "Product": fileid, "Filename": filenames}
df = pd.DataFrame(data)
```

```python
roofing_subtypes = df[df["Category"] == "waterproofing"]["Sub-Category"].unique()
len(set(roofing_subtypes))
```

```python
df.to_csv('categories_filenames_raw.csv', index=False)
```
